# Image Enhanchment basic operations

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("images/New_Zealand_Coast.jpg", cv2.IMREAD_COLOR)[:, :, ::-1]
window = cv2.imread("images/building-windows.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=[20, 8])
plt.subplot(131)
plt.imshow(image)
plt.subplot(132)
plt.imshow(window, cmap="gray")
```

### Addition / Subtraction - Brightness

```python
matrix = np.ones(image.shape, dtype=np.uint8) * 50

bright = cv2.add(image, matrix)
dark = cv2.subtract(image, matrix)

plt.figure(figsize=[20,10])

plt.subplot(141)
plt.title("Original")
plt.imshow(image)

plt.subplot(142)
plt.title("Bright")
plt.imshow(bright)

plt.subplot(143)
plt.title("Dark")
plt.imshow(dark)

plt.show()
```

### Multiplication / Division - Contrast

```python
matrix1 = np.ones(image.shape) * 1.3
matrix2 = np.ones(image.shape) * 0.7

high = np.uint(cv2.multiply(np.float64(image), matrix1))
low = np.uint(cv2.multiply(np.float64(image), matrix2))

plt.figure(figsize=[20,10])


plt.subplot(141)
plt.title("Brighter")
plt.imshow(high)

plt.subplot(142)
plt.title("Original")
plt.imshow(image)

plt.subplot(143)
plt.title("Darker")
plt.imshow(low)

plt.show()
```

### Threshold

```python
ret_val, thres = cv2.threshold(window, 100, 255, cv2.THRESH_BINARY)
print(ret_val)
plt.figure(figsize=[20,8])
plt.subplot(131)
plt.title("Threshold")
plt.imshow(thres, cmap="gray")
thres = cv2.adaptiveThreshold(window, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 5)
plt.subplot(132)
plt.title("Adaptive")
plt.imshow(thres, cmap="gray")
```

### Project 1: Music Sheet Reader

```python
img = cv2.imread("images/Piano_Sheet_Music.png", cv2.IMREAD_GRAYSCALE)

ret_val, minRes = cv2.threshold(img, 70, 255, cv2.THRESH_BINARY)
ret_val2, maxRes = cv2.threshold(img, 130, 255, cv2.THRESH_BINARY)
adRes = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 7)

plt.figure(figsize=[33,9])
plt.subplot(141)
plt.title("Global " + str(ret_val))
plt.imshow(minRes, cmap="gray")

plt.subplot(142)
plt.title("Global " + str(ret_val2))
plt.imshow(maxRes, cmap="gray")

plt.subplot(143)
plt.title("Adaptive")
plt.imshow(adRes, cmap="gray")
```
