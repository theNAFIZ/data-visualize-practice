# Image Enhanchment basic operations

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("images/New_Zealand_Coast.jpg", cv2.IMREAD_COLOR)[:, :, ::-1]

plt.imshow(image)
```

### Addition / Subtraction - Brightness

```python
matrix = np.ones(image.shape, dtype=np.uint8) * 50

bright = cv2.add(image, matrix)
dark = cv2.subtract(image, matrix)

plt.figure(figsize=[20,10])

plt.subplot(141)
plt.title("Original")
plt.imshow(image)

plt.subplot(142)
plt.title("Bright")
plt.imshow(bright)

plt.subplot(143)
plt.title("Dark")
plt.imshow(dark)

plt.show()
```

### Multiplication / Division - Contrast

```python
matrix1 = np.ones(image.shape) * 1.3
matrix2 = np.ones(image.shape) * 0.7

high = np.uint(cv2.multiply(np.float64(image), matrix1))
low = np.uint(cv2.multiply(np.float64(image), matrix2))

plt.figure(figsize=[20,10])


plt.subplot(141)
plt.title("Brighter")
plt.imshow(high)

plt.subplot(142)
plt.title("Original")
plt.imshow(image)

plt.subplot(143)
plt.title("Darker")
plt.imshow(low)

plt.show()
```
