```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
logo = cv2.imread("images/coca-cola-logo.png", cv2.IMREAD_COLOR)[:,:,::-1]
bg = cv2.imread("images/checkerboard_color.png")[:,:,::-1]
```

```python
logo_gs = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)
ret, msk= cv2.threshold(logo_gs, 127, 255, cv2.THRESH_BINARY)
```

```python
bg_resize = cv2.resize(bg, (700, 700))
plt.imshow(bg_resize)
```

```python
mask_inv = cv2.bitwise_not(msk)
plt.imshow(mask_inv, cmap="gray")
```

```python
res_bg = cv2.bitwise_and(bg_resize, bg_resize, mask=msk)
plt.imshow(res_bg)
```

```python
res_lg = cv2.bitwise_and(logo, logo, mask=mask_inv)
plt.imshow(res_lg)
```

```python
res = cv2.bitwise_or(res_lg, res_bg)
plt.imshow(res)
```
