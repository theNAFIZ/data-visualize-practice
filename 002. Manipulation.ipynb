# Basic Manipulation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("images/New_Zealand_Boat.jpg", cv2.IMREAD_UNCHANGED), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## Modifying pixels

```python
img_copy = img.copy()
img_copy[240:340,380:480] = [0,0,0] # pixel at height 239 to 339 width 379 to 479 color black
plt.imshow(img_copy)
```

## Cropping image

```python
boat = img[240:340, 385:485]
plt.imshow(boat)
```

## Resize image

```python
boat_2x = cv2.resize(boat, (800,800), interpolation=cv2.INTER_AREA)
cv2.imwrite("images/boat_2x.jpg", cv2.cvtColor(boat_2x, cv2.COLOR_RGB2BGR))
plt.imshow(boat_2x)
```

## Flipping Image

```python
hflip = cv2.flip(img, 1) # Flipcode 1 for horizontal shift
vflip = cv2.flip(img, 0) # Flipcode 0 for vertical shift
bflip = cv2.flip(img, -1) # Flipcode -1 for both shift

plt.figure(figsize=[40,10])

plt.subplot(141)
plt.title("Horizontal")
plt.imshow(hflip)

plt.subplot(142)
plt.title("Vertical")
plt.imshow(vflip)

plt.subplot(143)
plt.title("Both")
plt.imshow(bflip)

plt.subplot(144)
plt.title("Original")
plt.imshow(img)

plt.show()
```
