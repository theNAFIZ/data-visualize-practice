# Annotation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images/road-monument.jpg", cv2.IMREAD_COLOR)[:,:,::-1]
plt.imshow(image)
```

### Draw Lines

```python
img_line = image.copy()
cv2.line(img_line, (200,200), (460,200), (0,0, 255), 3)

plt.imshow(img_line)
```

### Draw Circles

```python
img_circle = image.copy()

cv2.circle(img_circle, (67, 155), 20, [255, 0, 0], 3, cv2.LINE_AA)
plt.imshow(img_circle)
```
