# Annotation

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("images/road-monument.jpg", cv2.IMREAD_COLOR)[:,:,::-1]
plt.imshow(image)
```

### Draw Lines

```python
img_line = image.copy()
cv2.line(img_line, (200,200), (460,200), (0,0, 255), 3)

plt.imshow(img_line)
```

### Draw Circles

```python
img_circle = image.copy()

cv2.circle(img_circle, (67, 155), 20, [255, 0, 0], 3, cv2.LINE_AA)
plt.imshow(img_circle)
```

### Draw Rectangles

```python
img_rect = image.copy()

cv2.rectangle(img_rect, (500,340), (550, 380), [0,255,0], 3, cv2.LINE_8, 0)
plt.imshow(img_rect)
```

### Add Text

```python
img_text = image.copy()

cv2.putText(img_text, "Text Annotation", (150, 50), cv2.FONT_HERSHEY_PLAIN, 2.4, [0,0,0], 2, cv2.LINE_8)
plt.imshow(img_text)
```

### All annotations in one image

```python
img_all = image.copy()

cv2.line(img_all, (200,200), (460,200), (0,0, 255), 3)
cv2.circle(img_all, (67, 155), 20, [255, 0, 0], 3, cv2.LINE_AA)
cv2.rectangle(img_all, (500,340), (550, 380), [0,255,0], 3, cv2.LINE_8, 0)
cv2.putText(img_all, "Text Annotation", (150, 50), cv2.FONT_HERSHEY_PLAIN, 2.4, [0,0,0], 2, cv2.LINE_8)

plt.imshow(img_all)
```
