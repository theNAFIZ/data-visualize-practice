```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
circle = cv2.imread("images/circle.jpg", cv2.IMREAD_COLOR)
plt.imshow(circle)
```

```python
red = np.ones(circle.shape, np.uint8)
multiplier = np.array([255, 0, 0], np.uint8)
red = red * multiplier
```

```python
red_circle = cv2.multiply(circle, red)
plt.imshow(red_circle)
```

```python
inv_circle = cv2.bitwise_not(circle)
plt.imshow(inv_circle)
```

```python
green = np.ones(circle.shape, np.uint8)
multiplier = np.array([0,255,0], np.uint8)

green = green * multiplier

green_bg = cv2.multiply(green, inv_circle)
plt.imshow(green_bg)
```

```python
flag = cv2.bitwise_or(red_circle, green_bg)

plt.imshow(flag)
```
