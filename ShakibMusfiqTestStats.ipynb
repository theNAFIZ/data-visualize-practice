```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
url = 'https://www.espncricinfo.com/player/shakib-al-hasan-56143/bowling-batting-stats'
html = pd.read_html('https://www.espncricinfo.com/player/shakib-al-hasan-56143/bowling-batting-stats')
df = pd.DataFrame(html[5])
```

```python
def yr(year):
  sp = year.split('year ')
  return sp[1]

a = []
for i in range(len(df)):
  a.append(yr(df['Unnamed: 0'][i]))
a
df["Year"] = a

p = []
for i in range(len(df)):
  p.append(int(df["HS"][i].replace("*","")))
df.HS = p
```

```python
plt.figure(figsize=(20,10))
plt.bar(df.Year, df.Runs, color='lightgreen', alpha=0.7)

# plt.plot(df.Year, df.Runs, color='lightgreen', alpha=0.8)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df.Year, df.Avg, color='yellow', alpha=1)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df.Year, df.HS, color='blue', alpha=0.3)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df.Year, df.SR, color='red', alpha=0.4)
plt.xticks(rotation=90)
plt.show()
```

```python
url2 = 'https://www.espncricinfo.com/player/mushfiqur-rahim-56029/bowling-batting-stats'
html2 = pd.read_html(url2)
df2 = pd.DataFrame(html2[5])
a = []
for i in range(len(df2)):
  a.append(yr(df2['Unnamed: 0'][i]))

df2["Year"] = a

p = []
for i in range(len(df2)):
  p.append(int(df2["HS"][i].replace("*","")))
df2.HS = p

p = []
for i in range(len(df2)):
  p.append(float(df2["Avg"][i].replace("-","0")))
df2.Avg = p
```

```python
plt.figure(figsize=(20,10))
plt.bar(df2.Year, df2.Runs, color='lightgreen', alpha=0.7)

# plt.plot(df.Year, df.Runs, color='lightgreen', alpha=0.8)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df2.Year, df2.SR, color='yellow', alpha=0.7)

# plt.plot(df.Year, df.Runs, color='lightgreen', alpha=0.8)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.bar(df2.Year, df2.HS, color='red', alpha=0.7)

# plt.plot(df.Year, df.Runs, color='lightgreen', alpha=0.8)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(df.Runs, labels=df.Year, autopct='%2.2f%%')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.fill_between(df2.Year, df2.Runs, color="blue", alpha=0.3)
plt.plot(df2.Year, df2.Runs, color="blue", alpha=1)

plt.fill_between(df.Year, df.Runs, color="red", alpha=0.3)
plt.plot(df.Year, df.Runs, color="red", alpha=1)

plt.legend(["",'Mushfiq', "",'Shakib'])
plt.xticks(rotation=90)
plt.title("Total runs in a year comparison")
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.fill_between(df2.Year, df2.Avg, color="blue", alpha=0.3)
plt.plot(df2.Year, df2.Avg, color="blue", alpha=1)

plt.fill_between(df.Year, df.Avg, color="red", alpha=0.3)
plt.plot(df.Year, df.Avg, color="red", alpha=1)

plt.legend(["",'Mushfiq', "",'Shakib'])
plt.xticks(rotation=90)
plt.title("Avg in a year comparison")
plt.show()
```
