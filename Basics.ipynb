```python
import cv2
import matplotlib.pyplot as plt
```

# Read Image
* Grayscale
* Color
* Default

```python
igray = cv2.imread("images/checkerboard_18x18.png", cv2.IMREAD_GRAYSCALE)
icolor = cv2.imread("images/coca-cola-logo.png", cv2.IMREAD_COLOR)
idef = cv2.imread("images/New_Zealand_Lake.jpg", cv2.IMREAD_UNCHANGED)
```

# Show Image

```python
plt.figure(figsize=[30,10])

plt.subplot(141)
plt.imshow(igray, cmap="gray")
plt.title("Gray Image")

plt.subplot(142)
plt.imshow(icolor)
plt.title("Color Image")

plt.subplot(143)
plt.imshow(idef)
plt.title("Default colorscale Image")
```

## Oops! CV2 reads image in bgr but plt shows it in rgb.
### So, we need to convert it.
* use cvtColor function
* use array manipulation img[:, :, ::-1]

```python
plt.figure(figsize=[20,10])
icolor = cv2.cvtColor(icolor, cv2.COLOR_BGR2RGB)
plt.subplot(141)
plt.imshow(icolor)
plt.title("Color Image")

idef = idef[:, :, ::-1]
plt.subplot(142)
plt.imshow(idef)
plt.title("Default colorscale Image")
```

# Splitting colors

### To rgb color space

```python
lake_bgr = cv2.imread("images/New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)
lake_rgb = lake_bgr[:,:,::-1]
# can also use lake_rgb = cv22.cvtColor(lake_bgr, cv2.COLOR_BGR2RGB)

r,g,b = cv2.split(lake_rgb)

plt.figure(figsize=[20,5])
plt.subplot(141)
plt.imshow(r, cmap="gray")
plt.title("red")
plt.subplot(142)
plt.imshow(g, cmap="gray")
plt.title("green")
plt.subplot(143)
plt.imshow(b, cmap="gray")
plt.title("blue")
plt.subplot(144)
plt.imshow(cv2.merge((r,g,b)))
plt.title("merged")

print(r.shape)
```

### To hsv color space

```python
cb = cv2.imread("images/coca-cola-logo.png", cv2.IMREAD_COLOR)
cb_hsv = cv2.cvtColor(cb, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(cb_hsv)

plt.figure(figsize=[20,5])
plt.subplot(141)
plt.imshow(h, cmap="gray")
plt.title("H channel")
plt.subplot(142)
plt.imshow(s, cmap="gray")
plt.title("S channel")
plt.subplot(143)
plt.imshow(v, cmap="gray")
plt.title("V channel")
plt.subplot(144)
cb = cv2.cvtColor(cb, cv2.COLOR_BGR2RGB)
plt.imshow(cb)
plt.title("merged")
```
